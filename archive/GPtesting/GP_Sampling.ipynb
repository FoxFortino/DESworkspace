```python
def kernel(x1, x2, var=1/100):
#     return x1 * x2
#     return np.minimum(x1, x2)
    return np.exp(-((x1 - x2)**2) / (2 * var))
    
x = np.arange(0, 1, 0.005)
n = x.shape[0]
x1, x2 = np.meshgrid(x, x)
C = kernel(x1, x2)
mu = np.zeros(n)
plt.imshow(C)

plots = []
for i in range(3):
    y = np.random.multivariate_normal(mu, C)
    plots.extend([x, y])
    
sigma = np.sqrt(np.diag(C))

plt.figure(figsize=(14, 7))
plt.xlim((0, 1))
plt.ylim((-5, 5))    
plt.plot(*plots)
plt.fill_between(x, 2*sigma, -2*sigma, alpha=0.9, color='gray')
plt.show()
```
