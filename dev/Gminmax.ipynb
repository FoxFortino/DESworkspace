```python
import os

import GPRutils
import DESutils
import vK2KGPR

import numpy as np
import astropy.units as u

import warnings
warnings.filterwarnings("ignore")
```

```python
exps = DESutils.bandDict["g"]
for expNum in exps[:10]:
    dataContainers = {}
    try:
        dC = GPRutils.loadFITS(f"/home/fortino/thesis/{expNum}/GPR.{expNum}.g.fits")
        dataContainers["        Std"] = dC
    except IndexError:
        pass
    
    try:
        dC = GPRutils.loadFITS(f"/home/fortino/test_N/{expNum}/GPR.{expNum}.g.fits")
        dataContainers["        New"] = dC
    except IndexError:
        pass
    
    try:
        dC = GPRutils.loadFITS(f"/home/fortino/Gminmax/GPR.{expNum}.g.fits")
        dataContainers["        New+"] = dC
    except IndexError:
        pass
    
    try:
        dC = GPRutils.loadFITS(f"/home/fortino/test_RC/{expNum}/GPR.{expNum}.gRC.fits")
        dataContainers["        RC"] = dC
    except IndexError:
        pass
    
    try:
        dC = GPRutils.loadFITS(f"/home/fortino/Gminmax/GPR.{expNum}.gRC.fits")
        dataContainers["        RC+"] = dC
    except IndexError:
        pass

    print(f"{expNum}, g")
    names = ["      Method", "Reduction", "xi0", "xif", "K Variance", "Outer Scale", "Diameter", "Wind X", "Wind Y"]
    header = "".join([f"{name:<15}" for name in names])
    print(header)
    for method, dC in dataContainers.items():
        vals = [method]
        
        xi0 = dC.header["xi0"]
        Xerr = dC.header["xi0_Xerr"]
        Yerr = dC.header["xi0_Yerr"]
        xi0err = np.sqrt(Xerr**2 + Yerr**2)
        vals.append(f"{xi0:.2f} ± {xi0err:.2f}")
        
        xif = dC.header["xif"]
        Xerr = dC.header["xif_Xerr"]
        Yerr = dC.header["xif_Yerr"]
        xiferr = np.sqrt(Xerr**2 + Yerr**2)
        vals.append(f"{xif:.2f} ± {xiferr:.2f}")
        
        vals.insert(1, f"{xi0/xif:.2f}")
        
        vals.extend(np.abs(np.around(dC.params, 7)))

        line = "".join([f"{param:<15}" for param in vals])
        print(line)
    print()
```
