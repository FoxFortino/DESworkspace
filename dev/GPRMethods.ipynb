```python
import GPRutils
import DESutils
import vK2KGPR

import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

Compare the New method and Std method (which should be equivalent) by fitting the data with the New methood with the final parameters obtained via the Std method. We should see very similar results that differ only because their validation sets are different.

The results of this test show that these two methods are probably equivalent. Now I should go and do a test where make a kernel but vary the validation set and see how different the results and if they agree with what I see here.

```python
exps = DESutils.findExpNums()
```

```python
expNum = exps[0]
dC_std = GPRutils.loadFITS(f"/home/fortino/thesis/{expNum}/GPR.{expNum}.*.fits")

dataC = GPRutils.dataContainer()
dataC.load(expNum)

GP = vK2KGPR.vonKarman2KernelGPR(dataC)

GP.fit(dC_std.fitCorrParams)
dataC.postFitCorr_sigmaClip(GP)

GP.fit(dC_std.params)
dataC.JackKnife(GP)

dataC.fitCorrParams = dC_std.fitCorrParams
dataC.saveFITS("/home/fortino")
dC_new = GPRutils.loadFITS(f"/home/fortino/*{expNum}*fits")

dC_std.summarize(noplots=True)
dC_new.summarize(noplots=True)
```
